```python
def km_sem_leitura(df):
    ls_empyt_rodovia_kms = [] # lista de kms sem leituras
    index_diff = addresses.index.difference(df.Id) # extrair kms sem dados de leitura
    ls_empyt_rodovia_kms.append(index_diff) # adiciona na lista
    df_empyt_rodovia_kms = pd.DataFrame(ls_empyt_rodovia_kms[0], columns = ["Id"])
    df_empyt_rodovia_kms = df_empyt_rodovia_kms.drop([0])
    return df_empyt_rodovia_kms
```
